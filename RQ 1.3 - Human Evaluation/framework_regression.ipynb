# Regression for RQ 1.3
In this notebook, regression on the human evaluation data from RQ 1.3 is performed. That way, we can see if the toxicity score and the rate of profanity have a meaningful relation with the severeness category. First, the severeness variable is mapped to the numbers 0, 1, 2, 3 to enable linear regression. Then, the linear regression is performed.

```python
import pandas as pd
import statsmodels.api as sm

file_path = 'output_for_regression.csv'
data = pd.read_csv(file_path)

data['Severeness_Encoded'] = data['Severeness'].map({'Non-toxic': 0,'Low': 1, 'Medium': 2, 'High': 3})
```

```python
data
```

```python
X = data[['Toxicity Score', 'Ratio of profanity']]
y = data['Severeness_Encoded']                     

X_with_const = sm.add_constant(X)

model = sm.OLS(y, X_with_const)
results = model.fit()

print(results.summary())
```
